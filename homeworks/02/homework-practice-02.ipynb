```python
def build_execution_time_graph(time_non_vectorized, time_vectorized, data_sizes, function_name):
    f, (ax2) = plt.subplots(1, 1, figsize=(13, 5))
    
    ax2.plot(data_sizes, time_non_vectorized, label=u"non_vectorized")
    ax2.plot(data_sizes, time_vectorized, label=u"vectorized")
    ax2.set_title(function_name + " execution time")
    ax2.set_xlabel(u"data size")
    ax2.set_ylabel(u"execution time, seconds")
    ax2.grid()
    ax2.legend()

    f.show()   
```

```python
import timeit

def calculate_execution_time(function, parameters):
    start_timer = timeit.default_timer()
    function(*parameters)
    return (timeit.default_timer() - start_timer)
```

```python
def calculate_ridge_regularization_coefficient():
    regression = lm.RidgeCV(
        alphas=np.log2(np.arange(1e-6, 1e6, 100)),
        store_cv_values=True,
    )
    regression.fit(attributes_train, target_train)
    print("RidgeCV regression:")
    print("regularization coefficient:", regression.alpha_)
    print("score on train sample:", regression.score(attributes_train, target_train))
    print("score on test sample:", regression.score(attributes_test, target_test))
calculate_ridge_regularization_coefficient()
```

Постройте график зависимости $R^2$ для обучающей и тестовой (на кросс-валидации) выборок в зависимости от значения гиперпараметра. Для Lasso также постройте график зависимости количества ненулевых весов.

```python
def draw_ridge_alpha_correlation():
    scores_train = []
    scores_test = []
    alpha_values = np.arange(1e-6, 1e6, 100)
    for alpha_value in alpha_values:
        regression = lm.Ridge(alpha=alpha_value)
        regression.fit(attributes_train, target_train)
        scores_train.append(regression.score(attributes_train, target_train))
        scores_test.append(regression.score(attributes_test, target_test))
        
    f, (ax2) = plt.subplots(1, 1, figsize=(13, 5))

    ax2.plot(alpha_values, scores_test, label=u"test score")
    ax2.plot(alpha_values, scores_train, label=u"train score")
    ax2.set_title(u"train and test score correlation by alpha value in Ridge regression")
    ax2.set_xlabel(u"alpha value")
    ax2.set_ylabel(u"score")
    ax2.grid()
    ax2.legend()

    f.show()    
draw_ridge_alpha_correlation()
```

```python
def draw_lasso_alpha_correlation():
    scores_train = []
    scores_test = []
    alpha_values = np.arange(1e-6, 50, 0.005)
    for alpha_value in alpha_values:
        regression = lm.Lasso(alpha=alpha_value)
        regression.fit(attributes_train, target_train)
        scores_train.append(regression.score(attributes_train, target_train))
        scores_test.append(regression.score(attributes_test, target_test))
        
    f, (ax2) = plt.subplots(1, 1, figsize=(13, 5))

    ax2.plot(alpha_values, scores_test, label=u"test score")
    ax2.plot(alpha_values, scores_train, label=u"train score")
    ax2.set_title(u"train and test score correlation by alpha value in Lasso regression")
    ax2.set_xlabel(u"alpha value")
    ax2.set_ylabel(u"score")
    ax2.grid()
    ax2.legend()

    f.show()    
draw_lasso_alpha_correlation()
```

```python
def draw_lasso_nonzero_weight_count_correlation():
    nonzero_weight_count = []
    alpha_values = np.arange(1e-6, 50, 0.005)
    for alpha_value in alpha_values:
        regression = lm.Lasso(alpha=alpha_value)
        regression.fit(attributes_train, target_train)
        nonzero_weight_count.append(sum(regression.coef_ > 0))
        
    f, (ax2) = plt.subplots(1, 1, figsize=(13, 5))

    ax2.plot(alpha_values, nonzero_weight_count, label=u"nonzero weight count")
    ax2.set_title(u"nonzero weight count correlation by alpha value in Lasso regression")
    ax2.set_xlabel(u"alpha value")
    ax2.set_ylabel(u"score")
    ax2.grid()
    ax2.legend()

    f.show()    
draw_lasso_nonzero_weight_count_correlation()
```

```python
def calculate_regressions_with_standart_scaler():
    scaler = pr.StandardScaler()
    scaled_attributes_train = scaler.fit_transform(attributes_train, target_train)
    scaled_attributes_test = scaler.fit_transform(attributes_test, target_test)
    
    linear_regression = lm.LinearRegression()
    linear_regression.fit(scaled_attributes_train, target_train)
    ridge_regression = lm.Ridge()
    ridge_regression.fit(scaled_attributes_train, target_train)
    lasso_regression = lm.Lasso()
    lasso_regression.fit(scaled_attributes_train, target_train)
    print("standart scaler:")
    print("scaled simple linear regression score:", linear_regression.score(scaled_attributes_test, target_test))
    print("scaled ridge regression score:", ridge_regression.score(scaled_attributes_test, target_test))
    print("scaled lasso regression score:", lasso_regression.score(scaled_attributes_test, target_test))
    
calculate_regressions_with_standart_scaler()
```

```python
def calculate_regressions_with_minmax_scaler():
    scaler = pr.MinMaxScaler()
    scaled_attributes_train = scaler.fit_transform(attributes_train, target_train)
    scaled_attributes_test = scaler.fit_transform(attributes_test, target_test)
    
    linear_regression = lm.LinearRegression()
    linear_regression.fit(scaled_attributes_train, target_train)
    ridge_regression = lm.Ridge()
    ridge_regression.fit(scaled_attributes_train, target_train)
    lasso_regression = lm.Lasso()
    lasso_regression.fit(scaled_attributes_train, target_train)
    print("minmax scaler:")
    print("scaled simple linear regression score:", linear_regression.score(scaled_attributes_test, target_test))
    print("scaled ridge regression score:", ridge_regression.score(scaled_attributes_test, target_test))
    print("scaled lasso regression score:", lasso_regression.score(scaled_attributes_test, target_test))
    
calculate_regressions_with_minmax_scaler()
```
