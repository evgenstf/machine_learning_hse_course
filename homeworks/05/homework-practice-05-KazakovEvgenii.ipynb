```python
def plot_surface(clf, X, y):
    plot_step = 0.01
    palette = sns.color_palette(n_colors=len(np.unique(y)))
    cmap = ListedColormap(palette)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.3)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, alpha=.7,
                edgecolors=np.array(palette)[y], linewidths=2)
```

```python
from sklearn import preprocessing
from sklearn.metrics import accuracy_score

def process_mushrooms_dataset():
    X_data = pd.read_csv('agaricus-lepiota.data', header=None)
    for column in X_data.columns:
        label_encoder = preprocessing.LabelEncoder()
        X_data[column] = label_encoder.fit_transform(X_data[column])
    Y_data = X_data[0]
    X_data = X_data.drop([0], 1)
    X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.3)
    decision_tree = DecisionTree(
        feature_types=list(('categorical ' * 22).split(' '))[:-1]
    )
    decision_tree.fit(X_train.values, Y_train.values)
    prediction = decision_tree.predict(X_test.values)
    
    print("accuracy_score:", accuracy_score(prediction, Y_test.values))

process_mushrooms_dataset()
```
